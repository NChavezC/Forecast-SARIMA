```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import yfinance as yf
```

```python
ticker = "BCH"
```

```python
df = yf.download(ticker)
df = df["2009-01-01":]
df
```

```python
df = df.asfreq('B', method='ffill')
```

```python
plt.figure(figsize=(12, 8))
plt.plot(df["Adj Close"])
plt.title(f"{ticker} Adjusted Close Price")
plt.xlabel('Year')
plt.ylabel('Price')
```

```python
from statsmodels.tsa.stattools import adfuller
def adf_test(series,title=''):
    """
    Pass in a time series and an optional title, returns an ADF report
    """
    print(f'Augmented Dickey-Fuller Test: {title}')
    result = adfuller(series.dropna(),autolag='AIC') # .dropna() handles differenced data
    
    labels = ['ADF test statistic','p-value','# lags used','# observations']
    out = pd.Series(result[0:4],index=labels)

    for key,val in result[4].items():
        out[f'critical value ({key})']=val
        
    print(out.to_string())          # .to_string() removes the line "dtype: float64"
    
    if result[1] <= 0.05:
        print("Strong evidence against the null hypothesis")
        print("Reject the null hypothesis")
        print("Data has no unit root and is stationary")
    else:
        print("Weak evidence against the null hypothesis")
        print("Fail to reject the null hypothesis")
        print("Data has a unit root and is non-stationary")
```

```python
adf_test(df['Adj Close'])
```

```python
q_80 = int(len(df) * .8)
```

```python
y_train = df['Adj Close'].iloc[:q_80]
y_test = df['Adj Close'].iloc[q_80:]
```

```python
y_train.plot(figsize = (12, 8))
y_test.plot()
```

```python
from pmdarima import auto_arima
from statsmodels.tsa.statespace.sarimax import SARIMAX
```

```python
model = auto_arima(y_train, trace = True, seasonal = True, m = 5)
```

Potential models are: 
- SARIMA(4,1,3)x(0,0,1,5)
- SARIMA(4,1,3)x(2,0,0,5) 
- SARIMA(4,1,3)x(0,0,2,5)
- SARIMA(4,1,3)x(0,0,0,5) with intercept

```python
model_list = {"order": [(4, 1, 3), (4, 1, 3), (4, 1 ,3), (4, 1 ,3)], "seasonal_order": [(0,0,0,5), (2,0,0,5), (0,0,2,5), (0,0,0,5)], "trend": ["n", "n", "n", "c"]}
```

```python
from statsmodels.tools.eval_measures import rmse
```

```python
metrics = pd.DataFrame(data = {"Order" : [], "Seasonal Order": [], "Trend": [], "AIC": [], "BIC" : [], "RMSE": []})
metric_rows = []
   
start = len(y_train)
end = len(y_train) + len(y_test) -1

for order, seasonal_order, trend in zip(model_list["order"], model_list["seasonal_order"], model_list["trend"]):
    model= SARIMAX(y_train, order = order, seasonal_order = seasonal_order, enforce_invertibility = True, trend = trend).fit()

    y_pred = model.predict(start, end)
    rmse_model = rmse(y_test, y_pred)
    
    metric_rows.append({"Order" : order, "Seasonal Order": seasonal_order, "Trend": trend, "AIC": model.aic, "BIC" : model.bic, "RMSE": rmse_model})
    
metrics = pd.concat([metrics, pd.DataFrame(metric_rows)], ignore_index=True)
```

```python
metrics = metrics.sort_values("RMSE")
metrics
```

The best model is SARIMA(0,1,3)x(2,0,2,12)

```python
selected_model = SARIMAX(y_train, order = metrics["Order"].iloc[0], seasonal_order = metrics["Seasonal Order"].iloc[0], enforce_invertibility = True, trend = metrics["Trend"].iloc[0]).fit()
y_hat_test = selected_model.predict(start, end).rename("Test Predictions")
plt.figure(figsize = (12, 8))
y_test.plot(legend = True)
y_hat_test.plot(legend = True)
```

```python
final_model =  SARIMAX(df["Adj Close"], order = metrics["Order"].iloc[0], seasonal_order = metrics["Seasonal Order"].iloc[0], enforce_invertibility = True, trend = metrics["Trend"].iloc[0]).fit()
final_model.summary()
```
